```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
plt.style.use("ggplot")
```

```python
df=pd.read_csv("movies.csv")
df.head()
```

```python
print("rows and columns: ",df.shape)
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=["budget_million_usd","revenue_million_usd"])
df["director"]=df["director"].fillna("unknown")
df["rating"]=df["rating"].fillna(df["rating"].mean())
df.isnull().sum()
```

```python
df.describe()
```

```python
#Finding the top 10 highest-rated movies and display them in a table

top10=df.sort_values(by="rating",ascending=False).head(10)
top10[["title","genre","rating","year"]]
```

```python
#Plotting a bar chart of average ratings by genre

avg_genre=df.groupby("genre")["rating"].mean().sort_values()

plt.figure(figsize=(10,5))

plt.bar(avg_genre.index, avg_genre.values,width=0.5)

plt.title("Average Rating by Genre")

plt.xlabel("Genre")

plt.ylabel("Average Rating")

plt.xticks(rotation=45)

plt.yticks(np.arange(0,8,0.5))

plt.show()
```

```python
#Plotting a histogram of movie release years

plt.figure(figsize=(8,5))

plt.hist(df["year"], bins=15)

plt.title("Distribution of Movie Release Years")

plt.xlabel("Year")

plt.ylabel("Number of Movies")

plt.tight_layout()

plt.show()
```
